## Normal Distribution

```python
# Generate a normal distribution
np.random.seed(42)
data = np.random.normal(loc=5, scale=10, size=1000)

# Calculate the mean
mean = np.mean(data)

# Plot the distribution using Seaborn
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')

# Mark the mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')

# Add legend and labels
plt.legend()
plt.title('Normal Distribution with Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Standard Normal Distribution

A standard normal distribution has a mean of 0 and standard deviation of 1.

```python
# Generate a standard normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate the mean and standard deviation
mean = round(np.mean(data))
print(f'Mean: {mean:0.2f}')
std_dev = round(np.std(data))
print(f'Standard deviation: {std_dev:0.2f}')

# Plot the distribution using Seaborn
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')

# Mark the mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')

# Mark one standard deviation below and above the mean
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=2, label=f'-1 Std Dev: {mean - std_dev:.2f}')
plt.axvline(mean + std_dev, color='blue', linestyle='dashed', linewidth=2, label=f'+1 Std Dev: {mean + std_dev:.2f}')

# Add legend and labels
plt.legend()
plt.title('Standard Normal Distribution with Mean and Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Empirical Rule

Empirical rule:

- 68% of the data falls within 1 standard deviation of the mean
- 95% of the data falls within 2 standard deviations of the mean
- 99.7% of the data falls within 3 standard deviations of the mean

For a standard normal distribution that means:

- 68% of the data falls between -1 and 1
- 95% of the data falls between -2 and 2
- 99.7% of the data falls between -3 and 3

## Percentiles

The nth percentile specifies how much of the data is below a certain point.

- 95th percentile is the point at which 95% of the data is below it.
- 50th percentile is the point at which 50% of the data is below it, or the median.
- 75th percentile minus the 25th percentile is the IQR.


---

## Z-scores

The **z-score** is the number of standard deviations a point is from the mean.  

The formula for the z-score for normally distributed data is given by:

$
z = \frac{x - \mu}{\sigma}
$

where:
- \($ z $\) is the z-score,
- \($ x $\) is the value,
- \($ \mu $\) is the mean of the population,
- \($ \sigma $\) is the standard deviation of the population.


A **z-table** is a table which shows the percentage of area under the distribution curve for a particular z-score.  

A point above the mean will have a positive z-score, whereas a point below the mean will have a negative z-score.  

There is a positive z-table, for points above the mean, and a negative z-table for points below the mean. 

The z-score is the percentile rank of the point used to calcuate it.

---

## Threshold Calculation

A threshold or cutoff is the value in the data set at which point some percentage of the data is either above or below.  

For instance in a standard normal distribution 1 is the value at which 84% of the data is below and 16% of the data is above.

To calcuate the threshold:

1. Look up the percentage of interest, P, in the z-table
2. From the table determine the z-score for that percentage
3. Multiple the z-score by the standard deviation and add to it the mean
4. The result is the value at which P percent of the data is below.
