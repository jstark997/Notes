## Frequency Table

Below is a frequency table of how many times the Summer Olympic Games have been hosted on each continent.

| Continent | Count |
|:----------|:--------:|

| Australia | 2 |

## Bar Graph

### Using matplotlib

```python
# Sample data
categories = ['Europe', 'North America', 'Asia', 'Australia', 'South America']
values = [16, 6, 3, 2, 1]

# Creating the bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Continent')
plt.ylabel('Number of Hostings')
plt.title('Summer Olympic Games Hostings by Continent')
plt.show()
```

### Using seaborn

```python
# Sample data
data = pd.DataFrame({
    'Continent': ['Europe', 'North America', 'Asia', 'Australia', 'South America'],
    'Hostings': [16, 6, 3, 2, 1]
})

# Creating the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Continent', y='Hostings', data=data, palette='Blues_d')
plt.xlabel('Continents')
plt.ylabel('Number of Hositings')
plt.title('Summer Olympic Games Hostings by Continent')
plt.show()
```

## Pie Chart

### Using matplotlib

```python
# Sample data
labels = ['Europe', 'North America', 'Asia', 'Australia', 'South America']
sizes = [16, 6, 3, 2, 1]
colors = ['red', 'yellow', 'green', 'blue', 'purple']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

# Custom function to display numbers
def absolute_value(val):
    a  = int(val/100.*sum(sizes))
    return f'{a}'

# Creating the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct=absolute_value, shadow=True)
plt.title('Summer Olympic Hostings by Continent')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Using seaborn

Note: Pie charts are not supported by seaborn directly. But can use seaborn color palette with matplotlib to create a pie chart.

```python
# Sample data
labels = ['Europe', 'North America', 'Asia', 'Australia', 'South America']
sizes = [16, 6, 3, 2, 1]
colors = sns.color_palette('pastel')[0:4]
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

# Custom function to display numbers
def absolute_value(val):
    a  = int(val/100.*sum(sizes))
    return f'{a}'

# Creating the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True)
plt.title('Summer Olympic Hostings by Continent')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
