## Left-Skewed (Negative) Distributions

```python
# Generate a left-tailed distribution
numValues = 1000
maxValue = 100
skewness = -5   #Negative values are left skewed, positive values are right skewed.

data = stats.skewnorm.rvs(a = skewness,loc=maxValue, size=numValues)  #Skewnorm function

data = data - min(data)      #Shift the set so the minimum value is equal to zero.
data = data / max(data)      #Standadize all the vlues between 0 and 1. 
data = data * maxValue         #Multiply the standardized values by the maximum value.
data = data.astype(int)

# Calculate the mode, median, and mean
count = Counter(data) 
get_mode = dict(count) 
mode_list = [k for k, v in get_mode.items() if v == max(list(count.values()))] 
print(mode_list[0])

has_mode = True if len(mode_list) == 1 else False

median = np.median(data)
mean = np.mean(data)

# Plot the distribution using Seaborn
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')

# Mark the mode, median, and mean
if has_mode:
    plt.axvline(mode_list[0], color='red', linestyle='dashed', linewidth=2, label=f'Mode: {mode_list[0]:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mean, color='blue', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')

# Add legend and labels
plt.legend()
plt.title('Left-Tailed Distribution with Mode, Median, and Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

In a left-skewed distribution the **mean** is less than the **median** which is less than the **mode**.

## Right-Skewed (Positive) Distributions

```python
# Generate a right-tailed distribution
numValues = 1000
maxValue = 100
skewness = 5   #Negative values are left skewed, positive values are right skewed.

data = stats.skewnorm.rvs(a = skewness,loc=maxValue, size=numValues)  #Skewnorm function

data = data - min(data)      #Shift the set so the minimum value is equal to zero.
data = data / max(data)      #Standadize all the vlues between 0 and 1. 
data = data * maxValue         #Multiply the standardized values by the maximum value.
data = data.astype(int)

# Calculate the mode, median, and mean
count = Counter(data) 
get_mode = dict(count) 
mode_list = [k for k, v in get_mode.items() if v == max(list(count.values()))]
print(mode_list[0])

has_mode = True if len(mode_list) == 1 else False

median = np.median(data)
mean = np.mean(data)

# Plot the distribution using Seaborn
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')

# Mark the mode, median, and mean
if has_mode:
    plt.axvline(mode_list[0], color='red', linestyle='dashed', linewidth=2, label=f'Mode: {mode_list[0]:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mean, color='blue', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')


# Add legend and labels
plt.legend()
plt.title('Right-Tailed Distribution with Mode, Median, and Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

In a right-skewed distribution the **mode** is less than the **median** which is less than the **mean**.
