# Figures And Multiple Plots

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Numpy

The **numpy** library is a Python library for performing mathematical opeartions on vectors and matrices.

We will be using this library as matplotlib can plot the contents of numpy arrays (vectors) and numpy makes it easy to generate and manipulate vectors, which is convenient for create data to plot.

```python
# Example of using numpy for data creation
x = np.arange(5)
y = np.arange(5) * np.arange(5)
print(x)
print(y)
```

```python
# Example of plotting numpy created data
plt.plot(x,y)
```

## Figures And Axes

The **figure** is the overall container for everything you want to draw. It's like the entire canvas where all plots, texts, legends, and axes are placed. A figure can contain multiple plots (or subplots). You can think of it as the space where various visualizations are organized.

**Axes** refers to the part of the figure that contains the actual plot, including elements like the x-axis, y-axis, and plotted data. Each axes object can hold one plot. If you have a figure with multiple plots (subplots), each plot will have its own axes. An axes object can contain plots of more than one set of data.

```python
# Create 3 graphs in the same figure using the same axes object
nums = np.arange(5)
plt.plot(nums,nums)
plt.plot(nums,nums*nums)
plt.plot(nums,nums**3)
```

To create a new figure call the **figure** method.

This method creates a new figure and activates it as the current figure and returns it as a Figure object.

When a figure is active, then all subsequent actions become a part of that figure, **until a new figure is created**.

The **plot** method uses the figure object that is currently active, otherwise if no figure object is active it will create one.

```python
# Create a new figure
plt.figure()
```

```python
# Add a plot to the figure above
plt.plot(nums,nums)
```

```python
# Create another figure and add plots to it
plt.figure()
plt.plot(nums,nums*nums)
plt.plot(nums,nums**3)
```
