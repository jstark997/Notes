# X And Y Ticks

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
```

## Setting Ticks

The methods **xticks** and **yticks** set the x and y axes ticks given an array of tick numbers and labels.

Tick numbers do not need to be evenly spaced.

Labels are specified with the **labels** parameter.

```python
# Change the x ticks for the age data of the age vs salary plot
plt.plot(ages, salaries)
plt.xticks([10,20,30,40,50])
```

```python
# Spacing of ticks does not need to be even
plt.plot(ages, salaries)
plt.xticks([20,30,37,40,50])
```

```python
# Set new labels for the y axis
plt.plot(ages, salaries)
plt.yticks([60000,80000,100000,120000,140000],labels=['60K','80K','100K','120K','140K'])
```

## Setting Axis Limits

The methods **xlim** and **ylim** take numbers as the start and end limits for the x and y axes respectively.

```python
# Set the x limits to 25 to 40 (effectively cutting of some of the data)
plt.plot(ages, salaries)
plt.xlim(25,40)
```

```python
# Set the y limits to 40000 to 180000
plt.plot(ages, salaries)
plt.ylim(40000,180000)
```
