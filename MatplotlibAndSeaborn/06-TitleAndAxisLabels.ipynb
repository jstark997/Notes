# Title And Axis Labels

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data points
nums = np.arange(5)

# More data
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
```

## Title

The **title** method will cause the passed string argument to be displayed as the title of the current plot.

The **loc** parameter can be used to change the position of the title: 'left','center','right' (default is 'center')

Some other parameters:

- labelpad - float
- color - CSS name or RGB hex code
- fontsize - float
- fontweight - float
- fontstyle - 'normal','italic','oblique')
- rotation - float, or 'vertical', 'horizontal'

```python
# Add a title to the age vs. salary plot
plt.plot(ages,salaries)
plt.title('Company Salaries By Age')
```

```python
# Add a title on the left for the simple linear function plot
plt.plot(nums,nums)
plt.title('Linear Function Example', loc='left')
```

## Axis Labels

The methods **xlabel** and **ylabel** will display the string argument passed along the x and y axes respectively.

Some parameters:

- labelpad - float
- color - CSS name or RGB hex code
- fontsize - float
- fontweight - float
- fontstyle - 'normal','italic','oblique')
- rotation - float, or 'vertical', 'horizontal'

```python
# Label the x and y axes
plt.plot(ages,salaries)
plt.title('Company Salaries By Age')
plt.xlabel('Employee Age')
plt.ylabel('Average Salary')
```
