# Setting Specific Style Attributes

The method **plot** has many parameters, such as 'color' (or 'c'), that can be used to set the style and change the appearance of (Line2D) plots.

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data points
nums = np.arange(5)
```

## Color

The **color** (or **c**) parameter of the plot method will set the color of a plot.

This parameters accepts:

- CSS named color
- RGB hex code

```python
plt.plot(nums,nums, color='purple')
plt.plot(nums,nums**2, color='#ff6b6b')
plt.plot(nums,nums**3, c='#ff9f43')
```

## Line Width

The **linewidth** (or **lw*) parameter takes a number (float) that sets the width of the line for the plot.



```python
plt.plot(nums,nums, color='purple', linewidth=3)
plt.plot(nums,nums**2, linewidth=5, color='#ff6b6b')
plt.plot(nums,nums**3, c='#ff9f43', lw=7)
```

## Line Style

The **linestyle** parameter takes a string argument that describes the style of the plot line.

Arguments:

- '-' or 'solid'
- '--' or 'dashed'
- '-.' or 'dashdot'
- ':' or 'dotted'
- '' or ' ', 'none', 'None' to draw nothing

```python
plt.plot(nums,nums, color='purple', linestyle='dashed')
plt.plot(nums,nums**2, color='#ff6b6b', linestyle='-.')
plt.plot(nums,nums**3, c='#ff9f43', linestyle=':')
```

## Marker

A marker is a shape that shows the point at which a plot crosses the y-axis through an x-axis label.

The **marker** parameter takes a string argument that describes the shape of the marker.

Some arguments:

- 'o' circle
- 's' square
- - 'x' x
- '+' plus
- '*' star
- '^' triangle up
- 'D' diamond
- 'd' thin diamond

The **markersize** parameter takes a number which sets the size of the marker.

The **markevery** parameter takes an int, list of int, list of boolean arguments to set how often a marker appears. Default is every other x-axis label.

The **markerfacecolor** parameter takes CSS name or RGB hex code to set the color of the face of the marker.

The **markeredgecolor** parameter takes CSS name or RGB hex code to set the color of the edge of the marker.

```python
plt.plot(nums,nums, color='purple', marker='o', markersize=20)
plt.plot(nums,nums**2, color='#ff6b6b', marker='x', markevery=2)
plt.plot(nums,nums**3, c='#ff9f43', marker='^', markerfacecolor='green', markeredgecolor='green')
```
