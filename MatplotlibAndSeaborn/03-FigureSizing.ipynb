# Figure Sizing

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Setting Figure Size

The parameter **figsize** of the figure method expects a tuple that specifies the width and height of the figure in 'inches', where an inch depends on the resolution of the display. The default figsize is width 6 inches and height 4 inches. Matplotlib will automatically scale the size of the figure so that it fits the available space.

The **dpi** parameter sets the resolution of the figure, where the default is 100.

```python
# Create an array of 5 numbers as data points
nums = np.arange(5)
```

```python
# Plot the data using the default figure size
plt.plot(nums,nums)
```

```python
# Create a new figure, defining the size as width 2 inches and height 6 inches and plot the data
plt.figure(figsize=(2,6))
plt.plot(nums,nums)
```

```python
# Create a new figure, defining the size as width 10 inches and height 6 inches and plot the data
plt.figure(figsize=(10,6))
plt.plot(nums,nums)
```

```python
# Set the dpi parameter to 200
plt.figure(dpi=200)
plt.plot(nums,nums)
```

```python
# Set the dpi parameter to 20
plt.figure(dpi=20)
plt.plot(nums,nums)
```
