## Show Versus Plot Methods

The method **show** renders the plot outside of an interactive environment (such as a Jupyter notebook). The **plot** method just returns a matplotlib object that some environments (such as Jupyter notebook) automatically render.

```python
# Create some data to plot
ages = [20,25,30,32,40,45]
salaries = [55000,60000,65000,72000,90000,115000]
```

```python
# Use only plot method
# This will return a Line2D object that Jupyter notebook automatically renders
plt.plot(ages,salaries)
```

```python
# Use show method
# The call to show is what renders the plot
plt.plot(ages,salaries)
plt.show()
```
