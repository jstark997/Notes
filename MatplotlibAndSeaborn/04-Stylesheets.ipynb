# Stylesheets

Matplotlib comes with many built in pre-configured styles for many different attributes available in stylesheets.

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data points
nums = np.arange(5)
```

## List Stylesheets

```python
plt.style.available
```

## Setting A Stylesheet

The method **style.use** will set the stylesheet to the the one named in the argument passed.

```python
# Set style to 'fivethirtyeight'
plt.style.use('fivethirtyeight')
plt.plot(nums,nums)
plt.plot(nums,nums**2)
plt.plot(nums,nums**3)
```

```python
# Set style to 'fivethirtyeight'
plt.style.use('seaborn-v0_8')
plt.plot(nums,nums)
plt.plot(nums,nums**2)
plt.plot(nums,nums**3)
```

```python
# Set style to 'fivethirtyeight'
plt.style.use('Solarize_Light2')
plt.plot(nums,nums)
plt.plot(nums,nums**2)
plt.plot(nums,nums**3)
```
