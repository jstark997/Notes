# What Is Matplotlib

The **matplotlib** library is a Python library for visualization that supports many different types of plots.

Some of the plots supported by matplotlib:

- Line
- Scatter
- Bar
- Pie
- Histogram
- Box and whiskers



## Setup

```python
import matplotlib.pyplot as plt
```

## Basic Plotting

The method **plot** will generate a plot of whatever data is passed to it. 

```python
# Empty plot
plt.plot()
```

```python
# Simple plot
# With only a single array, matplotlib treats those as the y-axis values,
# and makes up x-axis values starting at 0 and incrementing by 1
plt.plot([2,4,6,8])
```

```python
# Another simple plot example
plt.plot([2,4,2,8,6])
```

```python
salaries=[55000,65000,72000,90000,115000,150000]
ages=[20,25,30,32,40,45]
#plot ages vs salaries, where ages is on the x-axis and salaries is on the y-axis
plt.plot(ages, salaries)
```
