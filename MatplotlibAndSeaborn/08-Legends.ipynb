# Legends

## Setup

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
nums = np.arange(5)
```

## Creating A Legend

The **legend** method will create a legend for the lines in the plot. However, each line must be labeled (with the 'label' parameter).

Matplotlib will do its best to position the legend so that none of the lines cross over it. However, the 'loc' parameter can be used to specify a location for the legend.

Some parameters:

- frameon - True of False
- shadow - True or False
- facecolor - CSS name or RGB hex code

```python
plt.figure(figsize=(5,5))
plt.plot(nums,nums,c='teal',label='x')
plt.plot(nums,nums**2,c='olive',label='x squared')
plt.plot(nums,nums**3,c='purple',label='x cubed')
plt.legend()
```

```python
# Add a frame and shadow around the legend and change its face color to yellow
plt.figure(figsize=(5,5))
plt.plot(nums,nums,c='teal',label='x')
plt.plot(nums,nums**2,c='olive',label='x squared')
plt.plot(nums,nums**3,c='purple',label='x cubed')
plt.legend(frameon=True, shadow=True, facecolor='yellow')
```
